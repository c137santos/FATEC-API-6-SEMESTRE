```python
import geopandas as gpd
import pandas as pd
import os

gpd_path = 'ENEL_GO_6072_2023-04-30_V11_20240201-0941.gdb'

layer_name = 'SSDMT'
colunas_usadas = ['DIST', 'CTMT', 'CONJ', 'COMP', 'geometry']

print(f"Lendo a camada {layer_name}...")

if not os.path.exists(gpd_path):
    print(f"Erro: O arquivo '{gpd_path}' não foi encontrado. Por favor, verifique se o arquivo zip foi descompactado corretamente.")
    print("Arquivos disponíveis no diretório atual:")
    print(os.listdir('/content'))
else:
    gdf = gpd.read_file(gpd_path, layer=layer_name, columns=colunas_usadas, engine='pyogrio')

    gdf['COMP'] = pd.to_numeric(gdf['COMP'], errors='coerce')
    gdf['COMP_KM'] = gdf['COMP'] / 1000

    calculo_trechos = gdf.groupby(['DIST', 'CONJ', 'CTMT'])['COMP_KM'].sum().reset_index()

    print("\nResumo de Extensão de Rede MT (em Km):")
    print(calculo_trechos.head())
```
